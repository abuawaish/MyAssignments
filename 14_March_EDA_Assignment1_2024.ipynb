# <b><font color = 'orange'> About-Dataset </font></b>

- <font color = 'lightblue'><b><p>The dataset provided for this project is focused on heart failure, a critical medical condition characterized by the heart's inability to pump blood effectively, leading to inadequate circulation throughout the body. It comprises medical records of 299 patients diagnosed with heart failure, collected at the Faisalabad Institute of Cardiology and the Allied Hospital in Faisalabad, Punjab, Pakistan, during the period of April to December in 2015. The dataset includes various clinical, body, and lifestyle information of the patients, encompassing 13 features providing insights into the patients' health status and potential risk factors for heart failure. These features include age, gender, medical conditions such as diabetes and high blood pressure, as well as various medical measurements like platelet count, serum creatinine levels, and more. Additionally, the dataset indicates whether the patient died during the follow-up period.</font></p></b>

# <b><font color = 'orange'> Features </font></b>
- <b><font color = 'lightblue'>1. <u>`Age`</u> : The age of the patient (numeric).
- <b><font color = 'lightblue'>2. <u>`Anaemia`</u> : Indicates the presence of anemia (binary: 0 for no, 1 for yes).
- <b><font color = 'lightblue'>3. <u>`Creatinine Phosphokinase (CPK)`</u> : Level of creatinine phosphokinase enzyme in the blood (numeric).
- <b><font color = 'lightblue'>4. <u>`Diabetes`</u> : Indicates whether the patient has diabetes (binary: 0 for no, 1 for yes).
- <b><font color = 'lightblue'>5. <u>`Ejection Fraction`</u> : Percentage of blood leaving the heart at each contraction (numeric).
- <b><font color = 'lightblue'>6. <u>`High Blood Pressure`</u> : Indicates whether the patient has high blood pressure (binary: 0 for no, 1 for yes).
- <b><font color = 'lightblue'>7. <u>`Platelets`</u> : Platelet count in the blood (numeric).
- <b><font color = 'lightblue'>8. <u>`Serum Creatinine`</u> : Level of creatinine in the blood (numeric).
- <b><font color = 'lightblue'>9. <u>`Serum Sodium`</u> : Level of sodium in the blood (numeric).
- <b><font color = 'lightblue'>10. <u>`Sex`</u> : Gender of the patient (binary: 0 for female, 1 for male).
- <b><font color = 'lightblue'>11. <u>`Smoking`</u> : Indicates whether the patient smokes (binary: 0 for no, 1 for yes).
- <b><font color = 'lightblue'>12. <u>`Time`</u> : Follow-up period (in days) for the patient's condition (numeric).
- <b><font color = 'lightblue'>13. <u>`Death Event`</u> : Indicates whether the patient died during the follow-up period (binary: 0 for no, 1 for yes).

# <b><font color = 'orange'> 1. what is The distribution of age among heart failure patients in the dataset?

```python
# Calculate mean age
mean_age = df['age'].mean()
print("Mean age:", mean_age)

# Calculate median age
median_age = df['age'].median()
print("Median age:", median_age)

# Calculate mode age
mode_age = df['age'].mode()[0]
print("Mode age:", mode_age)

# Calculate range of ages
min_age = df['age'].min()
max_age = df['age'].max()
range_age = max_age - min_age
print("Range of ages:", range_age)
Heart_Failure_Patients = df.loc[df['DEATH_EVENT'] == 1 , 'age'].count()
print('The total number of heart failure patients by age is : ' , Heart_Failure_Patients)
```

```python
df['mean_age'] = df['age'].mean()
df['median_age'] = df['age'].median()
mean_age = df.mean_age.mean()
median_age = df.median_age.median()
```

```python
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Heart Failure Patients',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('age',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1, label = f'Mean Age: {mean_age:.2f}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=1, label=f'Median Age: {median_age:.2f}')
plt.legend(shadow=True,edgecolor='black')
plt.show()
```

```python
sns.barplot(df.groupby(df['DEATH_EVENT'])['age'].value_counts()[1] , palette = 'plasma')
plt.xticks(rotation = 90)
plt.title('Heart Failure Patients by Age',fontdict={'color':'blue','family':'arial','size':20})
plt.ylabel('Count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xlabel('Age',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'> 2. How does the death rate vary age?

```python
# Calculate the death rate for different age groups
death_rate_by_age = df.groupby(pd.cut(df['age'], bins=10))['DEATH_EVENT'].mean()

# Creating a bar plot to visualize the death rate variation with age
death_rate_by_age.plot(kind='bar', color='lightcoral')
plt.title('Death Rate Variation with Age',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Age Group',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(rotation=45)
plt.ylabel('Death Rate',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'> 3. What is the percentage of male and female patients in the dataset?

```python
#Calculate the percentage of male and female patients
gender_distribution = df['sex'].value_counts(normalize=True) * 100

# Print the percentage of male and female patients
print("Percentage of Male Patients : {:.2f}".format(gender_distribution[1]))
print("Percentage of Female Patients : {:.2f}".format(gender_distribution[0]))
```

# <b><font color = 'orange'> 4. how the platelet count vary among different age groups?

```python
# Define the age groups
age_bins = [0, 40, 60, 80, 100]  # Define the age ranges for grouping
age_labels = ['0-40', '41-60', '61-80', '81-100']  # Labels for the age groups
df['AgeGroup'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Visualize the platelet count distribution among different age groups using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='AgeGroup', y='platelets', data = df , palette = 'rainbow')
plt.title('Platelet Count Variation Among Different Age Groups',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Age Group',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Platelet Count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'> 5. Is there a correlation between creatinine and sodium levels in the blood

```python
# Calculate the correlation between creatinine and sodium levels
correlation = df['serum_creatinine'].corr(df['serum_sodium'])

print(f"The correlation between creatinine and sodium levels is : {correlation}") # Negative correlation
```

# <b><font color = 'orange'> 6. How does the prevalence of high blood pressure differ between male and female patients?

```python
prevalence_data = df.groupby(['sex', 'high_blood_pressure']).size().unstack()

# Plotting a bar chart to visualize the prevalence of high blood pressure in male and female patients
prevalence_data.plot(kind='bar')
plt.title('Prevalence of High Blood Pressure by Gender',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Gender (0: Female, 1: Male)',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Number of Patients',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'], rotation=0)
legend = plt.legend(title='High Blood Pressure', labels=['No', 'Yes'],shadow=True,edgecolor='black',labelcolor = ['blue','orange'])
plt.setp(legend.get_title(), color='green')
plt.show()
```

```python
sns.boxplot(df.groupby('sex')['high_blood_pressure'].value_counts().unstack())
plt.title('Prevalence of High Blood Pressure by Gender',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Sex',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Number of Patients',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=0)
plt.show()
```

# <b><font color = 'orange'> 7. What is the relationship between smoking habits and the occurrence of heart failure?

```python
smoking_counts = df['smoking'].value_counts()
heart_failure_counts = df.groupby('smoking')['DEATH_EVENT'].value_counts().unstack()

# Calculate percentages
percentages = (heart_failure_counts.T / heart_failure_counts.T.sum()).T * 100

# Visualize the relationship
plt.figure(figsize=(8, 6))
sns.barplot(x=percentages.index, y=percentages[1], data=percentages, palette='muted')
plt.title('Percentage of Heart Failure Occurrences by Smoking Habits',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Smoking Status',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Percentage of Heart Failure Occurrences',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'> 8. Are there any noticeable patterns in the distribution of death events across different age groups?

```python
# Create a histogram to show the distribution of death events among different age groups
plt.figure(figsize=(10, 6))
plt.hist([df[df['DEATH_EVENT'] == 0]['age'], df[df['DEATH_EVENT'] == 1]['age']], 
         bins=30, color=['skyblue', 'salmon'], label=['Survived', 'Deceased'])
plt.xlabel('Age',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.title('Distribution of Death Events by Age Group',fontdict={'color':'blue','family':'arial','size':20})
plt.legend(shadow=True,edgecolor='black')
plt.show()
```

```python
# Create a box plot to compare the distribution of death events in different age groups
plt.figure(figsize=(8,5))
sns.boxplot(x='DEATH_EVENT', y='age', data=df ,palette = 'icefire')
plt.title('Distribution of Death Events Across Age Groups',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Death Event (0: No, 1: Yes)',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Age',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'>9. Is there any significant difference in ejection fraction between patients with and without diabetes?

```python
# Compare the ejection fraction between patients with and without diabetes
sns.boxplot(x='diabetes', y='ejection_fraction', data = df , palette = 'rainbow')
plt.title('Ejection Fraction Comparison for Patients with and without Diabetes',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Diabetes (0: No, 1: Yes)',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Ejection Fraction',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'>10. How does the serum creatinine level vary between patients who survived and those who did not?

```python
# Analyze serum creatinine levels based on patient survival
sns.histplot(df[df['DEATH_EVENT'] == 0]['serum_creatinine'], color='blue', kde=True, label='Survived')
sns.histplot(df[df['DEATH_EVENT'] == 1]['serum_creatinine'], color='red', kde=True, label='Not Survived')
plt.title('Serum Creatinine Levels Based on Patient Survival',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Serum Creatinine',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.legend(shadow = True , edgecolor = 'black')
plt.show()
```
