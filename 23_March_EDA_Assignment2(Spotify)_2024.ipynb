# <b><font color = 'orange'>1. Identify the top 5 popular artists based on the mean popularity of their tracks. Show the mean popularity of tracks for the top 5 popular artists using BarPlot.

```python
#First, calculate the mean popularity of tracks for each artist
mean_popularity_by_artist = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

#Select the top 5 popular artists based on the mean popularity
top_5_popular_artists = mean_popularity_by_artist.head(5)

# Plot the mean popularity of tracks for the top 5 popular artists using a BarPlot
plt.figure(figsize=(10, 6))
top_5_popular_artists.plot(kind='bar', color='skyblue')
plt.title('Mean Popularity of Tracks for Top 5 Popular Artists' , fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Artist',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Mean Popularity',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(rotation = 45)
plt.show()
```

# <b><font color = 'orange'>2. Determine the top 5 popular songs based on their popularity ratings. Display the popularity ratings of the top 5 popular songs using BarPlot.

```python
top_5_popular_songs = df.loc[df['Popularity'] >= 90,['Track Name','Popularity']].sort_values(by = 'Popularity' , ascending = False)[:5]
```

# <b><font color = 'orange'>3. Find the top 5 trending genres based on the mean popularity of tracks within each genre. Visualize the mean popularity of tracks for the top 5 trending genres.

```python
# Calculate the mean popularity of tracks for each genre
genre_mean_popularity = df.groupby('Track Name')['Popularity'].mean().sort_values(ascending=False)

# Identify the top 5 trending genres
top_trending_genres = genre_mean_popularity.nlargest(5)

# Create a bar plot to visualize the mean popularity of tracks for the top 5 trending genres
plt.figure(figsize=(12, 6))
top_trending_genres.plot(kind='bar', color='skyblue')
plt.title('Top 5 Trending Genres Based on Mean Popularity of Tracks',fontdict={'color':'blue','family':'arial','size':20})
plt.ylabel('Mean Popularity of Tracks',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xlabel('Genre',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(rotation=45 , ticks=[0,1,2,3,4] , labels=['hip_hop','pop','rock','soul','jazz'] , color = 'r')
plt.tight_layout()
plt.show()
```

# <b><font color = 'orange'>4. Identify the top 5 longest songs among the tracks of the top 5 popular artists. Represent the duration of the top 5 longest songs among the tracks of the top 5 popular artists using BarPlot.

```python
# Identify the top 5 popular artists based on the mean popularity of their tracks
top_popular_artists = df.groupby('Artist')['Popularity'].mean().nlargest(5).index.tolist()

# Extract the tracks of the top 5 popular artists
top_tracks = df[df['Artist'].isin(top_popular_artists)]

# Calculate the duration of each track and identify the top 5 longest songs
top_tracks['Duration'] =   df['Duration (ms)'].apply(lambda data : data//1000)# Convert duration in ms to duration in s
top_5_longest_songs = top_tracks.nlargest(5, 'Duration')

# Create a bar plot to visualize the duration of the top 5 longest songs
plt.figure(figsize=(10, 6))
plt.bar(top_5_longest_songs['Track Name'], top_5_longest_songs['Duration'], color='lightcoral')
plt.xlabel('Track Name',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Duration (in seconds)',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.title('Top 5 Longest Songs Among the Tracks of the Top 5 Popular Artists',fontdict={'color':'blue','family':'arial','size':20})
plt.xticks(rotation=45)
plt.show()
```
