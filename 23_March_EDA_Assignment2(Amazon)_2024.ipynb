- <b> `review_title` : Short review
- <b> `review_content` : Long review
- <b> `img_link` : Image link of the product
- <b> `product_link` : Official website link of the product

## <b><font color = 'blue'>Let's divide the dataset into two parts 1.Numerical Feature 2.categorical Feature

```python
numerical_feature = [f for f in df.columns if df[f].dtype != 'O']
categorical_feature = [f for f in df.columns if df[f].dtype == 'O']
print('we have {} numerical_feature : {}'.format(len(numerical_feature),numerical_feature))
print('we have {} categorical_feature : {}'.format(len(categorical_feature),categorical_feature))
```

# <b><font color = 'orange'>1. What is the average rating for each product category?

```python
#Calculate the average rating for each product category
average_rating_per_category = df.groupby('category')['rating'].mean()

# Display the the average rating for each product category
print(average_rating_per_category)
```

# <b><font color = 'orange'>2. What are the top rating_count products by category?

```python
# Grouping by category and finding the top rating_count products within each category
top_rated_products_by_category = df.sort_values('rating_count', ascending=False).groupby('category').head(1)

# Displaying the top rated products by category
top_rating_count = top_rated_products_by_category[['product_name','category','rating_count']][:5]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = top_rating_count , x = 'product_name' , y = 'rating_count' , palette = 'icefire')
plt.title('top rated products by category',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('product_name',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('rating_count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'>3. What is the distribution of discounted prices vs. actual prices?

```python
percentage_discount = (df.actual_price - df.discounted_price) / df.actual_price * 100
```

```python
plt.figure(figsize=(10, 6))
plt.hist(x = percentage_discount , bins=20, color='skyblue')
plt.title('Distribution of Percentage Discounts',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Percentage Discount',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Frequency',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(10, 6))
sns.scatterplot(data = df , x = df['discount_percentage'] , y = df['actual_price'] , marker = 'o' ,
                c = np.arange(0,len(df['actual_price'])) ,  s =  np.arange(0,len(df['actual_price'])))
plt.title('Distribution of discounted prices vs. actual prices',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('discount_percentage',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('actual_price',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.show()
```

# <b><font color = 'orange'>5. What are the most popular product names?

```python
# Count the frequency of each unique product name
popular_product_names = df['product_name'].value_counts().sort_values(ascending=False)

# Get the top 5 most popular product names
top_popular_product_names = popular_product_names.head(5)

print("Top 5 Most Popular Product Names:")
print(top_popular_product_names)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(top_popular_product_names,palette='rainbow')
plt.title('top 5 most popular product names',fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('product_name',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('count',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(color = 'k')
plt.show()
```

# <b><font color = 'orange'>6. What are the most popular product keywords?

```python
from collections import Counter
# Combine all the product keywords from the 'about_product' column
keywords = ' '.join(df['about_product'].astype(str))

# Split the combined keywords into a list of words
keyword_list = keywords.split()

# Count the frequency of each keyword
keyword_freq = Counter(keyword_list)

# Find the most common keywords
most_common_keywords = keyword_freq.most_common(10)  # Change the number to display more or fewer top keywords

# Display the most common keywords
for i,j in enumerate(most_common_keywords):
    print(i,j)
```

# <b><font color = 'orange'>7. What are the most popular product reviews?

```python
# Group by review_title and count the number of reviews
popular_reviews = df.groupby('review_title').size().nlargest(5)

# Plot the most popular reviews
obj = popular_reviews.plot(kind='bar', figsize=(10, 6), title='Top 5 Most Popular Product Reviews' , color = 'skyblue')
plt.title(obj.get_title() ,fontdict={'color':'blue','family':'arial','size':20})
plt.xlabel('Review Title',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.ylabel('Number of Reviews',fontdict={'color':'black','family':'arial','size':15,'fontweight':'bold'})
plt.xticks(rotation = 360, ticks = [0,1,2,3,4] , labels = ['r1','r2','r3','r4','r5'])
plt.show()
```

# <b><font color = 'orange'>8. What is the correlation between discounted_price and rating?

```python
# Calculate the correlation between discounted_price and rating
correlation = df['discounted_price'].corr(df['rating']) 
# Display the correlation coefficient
print(f"The correlation between discounted_price and rating is: {correlation}")
# Positive correlation it means if 1 unit change in 'discounted_price' then there is 12% chance to 'rating' is also increase
```

# <b><font color = 'orange'>9. What are the Top 5 categories based on the highest ratings?

```python
# Calculate the average rating for each category
average_ratings = df.groupby('category')['rating'].mean().sort_values(ascending=False)

# Select the top 5 categories with the highest average ratings
top_categories = average_ratings.head(5)

# Display the top 5 categories based on the highest ratings
print(top_categories)
```
